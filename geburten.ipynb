```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

from pyaxis import pyaxis
import pandas as pd
import sys
```

```python
# load data from bfs
URL = 'https://www.bfs.admin.ch/bfsstatic/dam/assets/17664382/master'
```

```python
px = pyaxis.parse(URL, encoding='ISO-8859-2')
```

```python
raw = px['DATA']
raw.head()
```

```python
# rename 'Kanton (-) / Bezirk (>>) / Gemeinde (......)' to 'name_de'
raw = raw.rename(columns={'Kanton (-) / Bezirk (>>) / Gemeinde (......)': 'name_de'})
print(raw.shape)
raw.head()
```

```python
# filter for gemeinden
raw = raw.loc[raw['name_de'].str.contains(r'[0-6][0-9][0-9][0-9]')]
print(raw.shape)
raw.head()
```

```python
# convert values 'name_de' to id
raw['name_de'] = raw['name_de'].str.slice(6, 11)
print(raw.shape)
raw.head()
```

```python
metadata= {}
for col in ['Geschlecht des Kindes', 'Staatsangehörigkeit (Kategorie) des Kindes', 'Altersklasse der Mutter']:
    conversion_dict = {}
    options = set(raw[col])
    for i, option in enumerate(options):
        conversion_dict[option] = i
    raw[col] = raw[col].map(conversion_dict)
    metadata[col] = conversion_dict
```

```python
raw = raw.astype('int')
```

```python
# create new dataframe
df_11_11101 = pd.DataFrame()
```

```python
# fill in values
df_11_11101['spatialunit_id'] = raw['name_de']
df_11_11101['time_value'] = raw['Jahr']
df_11_11101['timeinfo_id'] = 1
df_11_11101['value'] = raw['DATA']
df_11_11101['dim1_value_id'] = raw['Geschlecht des Kindes']
df_11_11101['dim2_value_id'] = raw['Staatsangehörigkeit (Kategorie) des Kindes']
df_11_11101['dim3_value_id'] = raw['Altersklasse der Mutter']
df_11_11101['indicator_id'] = 11101
```

```python
df_11_11101.head()
```

```python
# export as csv
df_11_11101.to_csv('11101_CH.csv', index=False, na_rep='NA')
```
